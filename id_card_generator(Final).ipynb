```python
import os
import random
import datetime
import qrcode
import re
from PIL import Image, ImageDraw, ImageFont
```

```python
image = Image.new('RGB',(1000,900),(255,255,255))
draw = ImageDraw.Draw(image)
font = ImageFont.truetype('arial.ttf', size=45)
```

```python
def emp_id():
    (x,y) = (600,50)
    idno = random.randint(10000000,90000000)
    emp_id.id = str('Id' + str(idno))
    color = 'rgb(0,0,0)'
    font = ImageFont.truetype('arial.ttf',size=60)
    draw.text((x,y),emp_id.id,fill=color,font=font)


def company_name():
    (x,y) = (50,50)
    cn = input('Enter your company name: ')
    company_name.company = cn[0:].upper()
    #print(company)
    color = 'rgb(0,0,0)'
    font = ImageFont.truetype('arial.ttf',size=80)
    draw.text((x,y),company_name.company,fill=color,font=font)


def emp_name():
    (x,y) = (50,250)
    f = input('Enter your first name: ')
    l = input('Enter your last name: ')
    first = f[0].upper() + f[1:].lower()
    last = l[0].upper() + l[1:].lower()
    emp_name.name = first+ " " + last
    #print(name)
    color = 'rgb(0,0,0)'
    font = ImageFont.truetype('arial.ttf',size = 40)
    draw.text((x,y),emp_name.name,fill=color,font=font)


def take_phoneNo():
    p = input('Enter your 10 digit Phone Number: ')
    check_phoneNo(p)
    phone = p


def check_phoneNo(message):
    p = re.compile("(0/91)?[7-9][0-9]{9}")
    
    if p.match(message):
        (x,y) = (50,850)
        color = 'rgb(0,0,0)'
        font = ImageFont.truetype('arial.ttf',size = 40)
        check_phoneNo.phone = message
        draw.text((x,y),check_phoneNo.phone,fill=color,font=font)
    
    else:
        print("Invalid number. Please check and try again")
        take_phoneNo()


def emp_address():
    (x,y) = (50,750)
    message = input('Enter your current address: ')
    color = 'rgb(0,0,0)'
    font = ImageFont.truetype('arial.ttf',size = 40)
    emp_address.address = message
    draw.text((x,y),emp_address.address,fill=color,font=font)


def take_DOB():
    d = input("Enter you date of birth in 'dd/mm/yyyy' format")
    check_DOB(d)


def check_DOB(message):
    day,month,year = message.split('/')
    isValid = True
    
    try:
        datetime.datetime(int(year),int(month),int(day))
    except ValueError:
        isValid = False
        
    if(isValid):    
        (x,y) = (50,550)
        color = 'rgb(0,0,0)'
        font = ImageFont.truetype('arial.ttf',size = 40)
        check_DOB.dob = message
        draw.text((x,y),check_DOB.dob,fill=color,font=font)
    
    else:
        print("Wrong format.Please check and try again")
        take_DOB()    


def take_gender():
    g = input("Enter your gender. Male, Female or Other?")
    check_gender(g)


def check_gender(message):
    (x,y) = (50,350)
    g = ['male','female','other']
    
    if message.lower() in g:
        color = 'rgb(0,0,0)'
        font = ImageFont.truetype('arial.ttf',size = 40)
        check_gender.gender = message
        draw.text((x,y),check_gender.gender,fill=color,font=font)
    
    else:
        print("Invalid input. Please try again")
        take_gender()


def take_bloodGroup():
    b = input('Enter your Blood Group: ')
    check_bloodGroup(b)


def check_bloodGroup(message):
    bg = ['A','B','O','AB']
    signs = ['-','+']
    blood_groups = []
    
    for i in bg:
        for j in signs:
            blood_groups.append(i+j)
    
    if message.upper() in blood_groups:
        (x,y) = (50,650)
        color = 'rgb(0,0,0)'
        font = ImageFont.truetype('arial.ttf',size = 40)
        check_bloodGroup.blood = message.upper()
        draw.text((x,y),check_bloodGroup.blood,fill=color,font=font)
    
    else:
        print("Wrong input. Please try again")
        take_bloodGroup()


def gen_image():
    image.save(str(emp_name.name)+'.png')
    img = qrcode.make(str(company_name.company)+str(emp_id.id))
    img.save(str(emp_id.id)+'.bmp')
    til = Image.open(emp_name.name+'.png')
    im = Image.open(str(emp_id.id)+'.bmp')
    til.paste(im,(600,350))
    til.save(emp_name.name+'.png')
    print("Your ID card has been saved")


def generate_card():
    emp_id()
    company_name()
    emp_name()
    take_phoneNo()
    emp_address()
    take_DOB()
    take_gender()
    take_bloodGroup()
    gen_image()
```

```python
i = input("Hello!! \nWould you like to generate your id card?\nType 'yes' or 'no': ")

if i[0].lower() == "y":
    generate_card()
    exit()

elif i[0].lower() == "n":
    print("You can close the window.")
    exit()

else:
    exit()
```
